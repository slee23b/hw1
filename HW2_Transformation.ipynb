```python
import pandas as pd

url = "https://raw.githubusercontent.com/slee23b/hw1/main/Customer%20Loyalty%20History.csv"
data = pd.read_csv(url)

data.head()
```

```python
#no date column in data
#other ways to transform data is removing null values and duplicates


#remove null
data.dropna(inplace=True)

#remove duplicates
data.drop_duplicates(inplace=True)

#new data
print("Transformed data:")
print(data.head())
```
